```python
#Sprawozdanie z eliminacji braków danych dla pliku "zad2_Stroke.csv"
import pandas as pd
from IPython.display import display
df = pd.read_csv ("zad2_Stroke.csv", sep=';')
```

```python
#Skonwertowanie zmiennych z typu object na typ float64
df['age'] = pd.to_numeric(df['age'].str.replace(',', '.'))
df['avg_glucose_level'] = pd.to_numeric(df['avg_glucose_level'].str.replace(',', '.'))
df['bmi'] = pd.to_numeric(df['bmi'].str.replace(',', '.'))
```

```python
df.dtypes
```

```python
df[['heart_disease', 'avg_glucose_level', 'bmi']].describe()
```

```python
#Eliminacje danych z wartością NaN
df.dropna()
```

```python
df[['heart_disease', 'avg_glucose_level', 'bmi']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' na średnie wartości
df['heart_disease'].fillna(df['heart_disease'].mean(), inplace=True)
df['avg_glucose_level'].fillna(df['avg_glucose_level'].mean(), inplace=True)
df['bmi'].fillna(df['bmi'].mean(), inplace=True)
```

```python
df[['heart_disease', 'avg_glucose_level', 'bmi']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' na medianę wartości
df['heart_disease'].fillna(df['heart_disease'].median(), inplace=True)
df['avg_glucose_level'].fillna(df['avg_glucose_level'].median(), inplace=True)
df['bmi'].fillna(df['bmi'].median(), inplace=True)
```

```python
df[['heart_disease', 'avg_glucose_level', 'bmi']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' za pomocą interpolacji liniowej
df['heart_disease'].fillna(df['heart_disease'].interpolate(), inplace=True)
df['avg_glucose_level'].fillna(df['avg_glucose_level'].interpolate(), inplace=True)
df['bmi'].fillna(df['bmi'].interpolate(), inplace=True)
```

```python
df[['heart_disease', 'avg_glucose_level', 'bmi']].describe()
```

```python
#Opis dla zmiennych kategorycznych
df[['work_type']].describe()
```

```python
#Zamiana zmiennych 'NaN' w zmiennych kategorycznych na zmienne najczęściej występujące
most_frequent = df['work_type'].mode().iloc[0]
df['work_type'].fillna(most_frequent, inplace=True)
```

```python
df[['work_type']].describe()
```

```python
##Druga część zadania polega na wykorzystaniu różnych prostych algorytmów uczenia maszynowego 
##do uzupełnienia luk brakującymi danymi
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych numerycznych
#Algorytm KNN
from sklearn.impute import KNNImputer
```

```python
columns_to_impute = ['heart_disease', 'avg_glucose_level', 'bmi']
data_to_impute = df[columns_to_impute].values
knn_imputer = KNNImputer(n_neighbors=16)
imputed_data = knn_imputer.fit_transform(data_to_impute)
df[columns_to_impute] = imputed_data
```

```python
df[['heart_disease', 'avg_glucose_level', 'bmi']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Algorytm KNN dla zmiennych kategorycznych
from sklearn.impute import SimpleImputer
```

```python
columns_to_impute = ['work_type']
imputer = SimpleImputer(strategy='most_frequent')
data_to_impute = df[columns_to_impute].values
imputed_data = imputer.fit_transform(data_to_impute)
df[columns_to_impute] = imputed_data
```

```python
df[['work_type']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Algorytm regresji logistycznej dla zmiennych numerycznych
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
X = df[['hypertension', 'stroke']]
y = df['work_type']
X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)

predicted_work= classifier.predict(X_test)

df.loc[df['work_type'].isnull(), 'work_type'] = predicted_work
```

```python
df[['work_type']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Algorytm regresji liniowej
from sklearn.linear_model import LinearRegression
```

```python
X = df[['age', 'hypertension']]
y = df['heart_disease']
X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)

predicted_heart= classifier.predict(X_test)

df.loc[df['heart_disease'].isnull(), 'heart_disease'] = predicted_heart
```

```python
X = df[['age', 'hypertension']]
y = df['avg_glucose_level']

X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

regressor = LinearRegression()
regressor.fit(X_train, y_train)

predicted_avg = regressor.predict(X_test)

df.loc[df['avg_glucose_level'].isnull(), 'avg_glucose_level'] = predicted_avg
```

```python
X = df[['age', 'hypertension']]
y = df['bmi']

X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

regressor = LinearRegression()
regressor.fit(X_train, y_train)

predicted_bmi = regressor.predict(X_test)

df.loc[df['bmi'].isnull(), 'bmi'] = predicted_bmi
```
