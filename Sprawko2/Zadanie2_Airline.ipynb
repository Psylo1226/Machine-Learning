```python
#Eliminacja danych z wartością NaN
df.dropna()
```

```python
#Statystyki po usunięciu wartości 'NaN' z kolumn
df[['Age', 'Gate.location', 'Arrival.Delay.in.Minutes']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' na średnie wartości
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Gate.location'].fillna(df['Gate.location'].mean(), inplace=True)
df['Arrival.Delay.in.Minutes'].fillna(df['Arrival.Delay.in.Minutes'].mean(), inplace=True)
```

```python
df[['Age', 'Gate.location', 'Arrival.Delay.in.Minutes']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' na medianę wartości
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Gate.location'].fillna(df['Gate.location'].median(), inplace=True)
df['Arrival.Delay.in.Minutes'].fillna(df['Arrival.Delay.in.Minutes'].median(), inplace=True)
```

```python
df[['Age', 'Gate.location', 'Arrival.Delay.in.Minutes']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' przy pomocy interpolacji liniowej
df['Age'].fillna(df['Age'].interpolate(), inplace=True)
df['Gate.location'].fillna(df['Gate.location'].interpolate(), inplace=True)
df['Arrival.Delay.in.Minutes'].fillna(df['Arrival.Delay.in.Minutes'].interpolate(), inplace=True)
```

```python
df[['Age', 'Gate.location', 'Arrival.Delay.in.Minutes']].describe()
```

```python
#Opis dla zmiennych kategorycznych
df[['Customer.Type']].describe()
```

```python
#Zamiana zmiennych 'NaN' w zmiennych kategorycznych na zmienne najczęściej występujące
most_frequent = df['Customer.Type'].mode().iloc[0]
df['Customer.Type'].fillna(most_frequent, inplace=True)
```

```python
df[['Customer.Type']].describe()
```

```python
##Druga część zadania polega na wykorzystaniu różnych prostych algorytmów uczenia maszynowego
##do uzupełniania luk brakującymi danymi
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić typ danych dla zmiennych numerycznych
#Algorytm KNN
from sklearn.impute import KNNImputer
```

```python
columns_to_impute = ['Age', 'Gate.location', 'Arrival.Delay.in.Minutes']
data_to_impute = df[columns_to_impute].values
knn_imputer = KNNImputer(n_neighbors=25)
imputed_data = knn_imputer.fit_transform(data_to_impute)
df[columns_to_impute] = imputed_data
```

```python
df[['Age', 'Gate.location', 'Arrival.Delay.in.Minutes']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Algorytm KNN dla zmiennych kategorycznych
from sklearn.impute import SimpleImputer
```

```python
columns_to_impute = ['Customer.Type']
imputer = SimpleImputer(strategy='most_frequent')
data_to_impute = df[columns_to_impute].values
imputed_data = imputer.fit_transform(data_to_impute)
df[columns_to_impute] = imputed_data
```

```python
df[['Customer.Type']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego
#Algorytm regresji logistycznej dla zmiennych numerycznych
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
X = df[['Flight.Distance', 'Seat.comfort']]
y = df['Customer.Type']
X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)

predicted_customer = classifier.predict(X_test)

df.loc[df['Customer.Type'].isnull(), 'Customer.Type'] = predicted_customer
```

```python
df[['Customer.Type']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego
#Algorytm regresji liniowej
from sklearn.linear_model import LinearRegression
```

```python
X = df[['Flight.Distance', 'Seat.comfort']]
y = df['Age']

X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

regressor = LinearRegression()
regressor.fit(X_train, y_train)

predicted_age = regressor.predict(X_test)

df.loc[df['Age'].isnull(), 'Age'] = predicted_age
```

```python
X2 = df[['Flight.Distance', 'Seat.comfort']]
y2 = df['Gate.location']

X2_train = X2[~y2.isnull()]
y2_train = y2.dropna()

X2_test = X2[y2.isnull()]

regressor = LinearRegression()
regressor.fit(X2_train, y2_train)

predicted_location = regressor.predict(X2_test)

df.loc[df['Gate.location'].isnull(), 'Gate.location'] = predicted_location
```

```python
X3 = df[['Flight.Distance', 'Seat.comfort']]
y3 = df['Arrival.Delay.in.Minutes']

X3_train = X3[~y3.isnull()]
y3_train = y3.dropna()

X3_test = X3[y3.isnull()]

regressor = LinearRegression()
regressor.fit(X3_train, y3_train)

predicted_arrival = regressor.predict(X3_test)

df.loc[df['Arrival.Delay.in.Minutes'].isnull(), 'Arrival.Delay.in.Minutes'] = predicted_arrival
```
