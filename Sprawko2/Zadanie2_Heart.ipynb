```python
#Sprawozdanie z eliminacji braków danych dla pliku "zad2_Heart.csv"
import pandas as pd
from IPython.display import display
df = pd.read_csv ("zad2_Heart.csv", sep=';')
```

```python
#Skonwertowanie zmiennych z typu object na typ float64
df['Oldpeak'] = pd.to_numeric(df['Oldpeak'].str.replace(',', '.'))
```

```python
df.dtypes
```

```python
df[['Age', 'Cholesterol', 'Oldpeak']].describe()
```

```python
#Eliminacja danych z wartością NaN
df.dropna()
```

```python
df[['Age', 'Cholesterol', 'Oldpeak']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' na średnie wartości
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Cholesterol'].fillna(df['Cholesterol'].mean(), inplace=True)
df['Oldpeak'].fillna(df['Oldpeak'].mean(), inplace=True)
```

```python
df[['Age', 'Cholesterol', 'Oldpeak']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' na medianę wartości
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Cholesterol'].fillna(df['Cholesterol'].median(), inplace=True)
df['Oldpeak'].fillna(df['Oldpeak'].median(), inplace=True)
```

```python
df[['Age', 'Cholesterol', 'Oldpeak']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Dla zmiennych numerycznych zamiana zmiennych 'NaN' przy pomocy interpolacji liniowej
df['Age'].fillna(df['Age'].interpolate(), inplace=True)
df['Cholesterol'].fillna(df['Cholesterol'].interpolate(), inplace=True)
df['Oldpeak'].fillna(df['Oldpeak'].interpolate(), inplace=True)
```

```python
df[['Age', 'Cholesterol', 'Oldpeak']].describe()
```

```python
#Opis dla zmiennych kategorycznych
df[['RestingECG', 'ExerciseAngina']].describe()
```

```python
#Zamiana zmiennych 'NaN' w zmiennych kategorycznych na zmienne najczęściej występujące
most_frequent1 = df['RestingECG'].mode().iloc[0]
df['RestingECG'].fillna(most_frequent1, inplace=True)
most_frequent2 = df['ExerciseAngina'].mode().iloc[0]
df['ExerciseAngina'].fillna(most_frequent2, inplace=True)
```

```python
df[['RestingECG', 'ExerciseAngina']].describe()
```

```python
##Druga część zadania polega na wykorzystaniu różnych prostych algorytmów uczenia maszynowego 
##do uzupełnienia luk brakującymi danymi
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Algorytm KNN
from sklearn.impute import KNNImputer
```

```python
columns_to_impute = ['Age', 'Cholesterol', 'Oldpeak']
data_to_impute = df[columns_to_impute].values
knn_imputer = KNNImputer(n_neighbors=19)
imputed_data = knn_imputer.fit_transform(data_to_impute)
df[columns_to_impute] = imputed_data
```

```python
df[['Age', 'Cholesterol', 'Oldpeak']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych numerycznych
#Algorytm KNN dla zmiennych kategorycznych
from sklearn.impute import SimpleImputer
```

```python
columns_to_impute = ['RestingECG', 'ExerciseAngina']
imputer = SimpleImputer(strategy='most_frequent')
data_to_impute = df[columns_to_impute].values
imputed_data = imputer.fit_transform(data_to_impute)
df[columns_to_impute] = imputed_data
```

```python
df[['RestingECG', 'ExerciseAngina']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Algorytm regresji logistycznej dla zmiennych numerycznych
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
X = df[['FastingBS', 'HeartDisease']]
y = df['ExerciseAngina']
X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)

predicted_resting= classifier.predict(X_test)

df.loc[df['ExerciseAngina'].isnull(), 'ExerciseAngina'] = predicted_resting
```

```python
df[['ExerciseAngina']].describe()
```

```python
##Należy najpierw zresetować dane do stanu początkowego i ustawić odpowiedni typ danych dla zmiennych
#Algorytm regresji liniowej
from sklearn.linear_model import LinearRegression
```

```python
X = df[['FastingBS', 'MaxHR']]
y = df['Age']

X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

regressor = LinearRegression()
regressor.fit(X_train, y_train)

predicted_age = regressor.predict(X_test)

df.loc[df['Age'].isnull(), 'Age'] = predicted_age
```

```python
X = df[['FastingBS', 'MaxHR']]
y = df['Cholesterol']

X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

regressor = LinearRegression()
regressor.fit(X_train, y_train)

predicted_cholesterol = regressor.predict(X_test)

df.loc[df['Cholesterol'].isnull(), 'Cholesterol'] = predicted_cholesterol
```

```python
X = df[['FastingBS', 'MaxHR']]
y = df['Oldpeak']

X_train = X[~y.isnull()]
y_train = y.dropna()

X_test = X[y.isnull()]

regressor = LinearRegression()
regressor.fit(X_train, y_train)

predicted_oldpeak = regressor.predict(X_test)

df.loc[df['Oldpeak'].isnull(), 'Oldpeak'] = predicted_oldpeak
```
